# **Introdução**


Python é uma linguagem de programação altamente flexível e poderosa que permite aos desenvolvedores manipular dados de diversas maneiras. Para que isso seja possível, Python utiliza variáveis e tipos de dados para armazenar e representar informações de diferentes formas. Vamos explorar os principais tipos de variáveis e tipos de dados em Python, com foco nos tópicos variáveis, números, strings e booleanos.

**Variáveis**:

Em Python, as variáveis são usadas para armazenar e manipular dados. Uma variável é essencialmente um nome que você atribui a um valor. Elas são cruciais para o funcionamento da linguagem, pois permitem que você armazene informações para uso posterior. As variáveis em Python podem conter uma variedade de tipos de dados, incluindo números, strings e booleanos.

**Números**:

Python suporta vários tipos de números, sendo os principais:

Inteiros (int): Representam números inteiros, positivos ou negativos, como 1, -42, 1000.
Números de ponto flutuante (float): Representam números com casas decimais, como 3.14, -0.001, 2.0.
Números complexos (complex): São usados para representar números complexos, como 2 + 3j, onde 'j' representa a parte imaginária.

**Strings**: 

São usadas para representar texto em Python. Elas são uma sequência de caracteres, como palavras, frases ou até mesmo caracteres individuais. Strings são definidas entre aspas simples ('') ou duplas (""). Por exemplo, "Python é incrível!" é uma string válida em Python.

**Booleanos**:

Os valores booleanos são usados para representar a verdade (True) ou a falsidade (False) de uma afirmação. Eles são fundamentais para a lógica de controle de fluxo em programação, permitindo que você tome decisões com base em condições. Por exemplo, em uma expressão condicional, você pode verificar se uma variável é igual a um valor específico e, com base nisso, tomar diferentes caminhos no seu programa.

Em resumo, em Python, variáveis são utilizadas para armazenar informações e podem conter números inteiros, números de ponto flutuante, strings e valores booleanos. Entender esses tipos de dados e como usá-los é essencial para escrever programas eficazes e flexíveis em Python.
